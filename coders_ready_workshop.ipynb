**Basemap**     : Basemap is a Python library that provides a set of tools for creating maps. It is built on top of the matplotlib library and provides a high-level interface for drawing maps and plotting data on them. Basemap supports several map projections and can draw coastlines, rivers, and political boundaries. It also supports adding annotations, such as text labels and scale bars, to the maps.

**Pandas**      : Pandas is a popular open-source data analysis and manipulation library for Python. It provides fast, flexible, and expressive data structures designed to work with both structured and unstructured data. Pandas is built on top of NumPy, another popular library for scientific computing in Python.

**Matplotlib**  : Matplotlib is a data visualization library in Python that provides a variety of tools for creating high-quality visualizations. It is built on NumPy and provides an easy-to-use interface for creating static, animated, and interactive plots.

```python
import pandas as pd
import plotly.express as pr
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/sample_data/airline_data.csv') #here df is data frame
df
```

```python
df.drop(['Unnamed: 0'], axis = 1) 
```

```python
df.isna().sum()
```

```python
line_data = df.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
line_data
```

### **graphical visuallisation**

```python
fig = go.Figure(data=go.Scatter(x=line_data['Month'], y=line_data['ArrDelay'], mode='lines', marker=dict(color='green')))
fig.update_layout(title='Monthly average delay of flights', xaxis_title='Months', yaxis_title ='Delay of the flight arrival')
fig.show()
```

### pi chart visuallisation

```python
import matplotlib.pyplot as plt
fig = pr.pie(df, names='DistanceGroup', title='Monthly Distance group propotion travelled')
##plt.savefig("pi.jpg") // here we can use any format like jpg, jped and png to save the visuallisations
fig.show()
```
